```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
train=pd.read_csv('/content/train_preprocessed.csv')
test=pd.read_csv('/content/test_preprocessed.csv')
```

```
train.head()
```

```
test.head()
```

```
X=train.to_numpy()
Z=test.to_numpy()
```

```
X_train, Y_train = X[:,2:], X[:,1]
X_test = Z[:,1:]
```

##Random Forest

```
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```
forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
forest_classifier.fit(X_train, Y_train)
```

```
forest_scores = cross_val_score(forest_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(forest_scores.mean())
```

##Logistic Regression

```
from sklearn.linear_model import LogisticRegression
```

```
logistic_classifier = LogisticRegression(random_state=42)
logistic_classifier.fit(X_train, Y_train)
```

```
logistic_scores = cross_val_score(logistic_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(logistic_scores.mean())
```

##Support Vector Machines

###Linear Kernel

```
from sklearn.svm import SVC
```

```
SVM_classifier = SVC(random_state=42, kernel='linear')
SVM_classifier.fit(X_train, Y_train)
```

```
SVM_scores = cross_val_score(SVM_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(SVM_scores.mean())
```

###Polynomial Kernel

```
SVM_classifier = SVC(random_state=42, kernel='poly', degree=3, coef0=1, C=5)
SVM_classifier.fit(X_train, Y_train)
```

```
SVM_scores = cross_val_score(SVM_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(SVM_scores.mean())
```

###Gaussian RBF Kernel

```
SVM_classifier = SVC(random_state=42)
SVM_classifier.fit(X_train, Y_train)
```

```
SVM_scores = cross_val_score(SVM_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(SVM_scores.mean())
```

##KNN Classifier

```
from sklearn.neighbors import KNeighborsClassifier
```

```
KNN_classifier = KNeighborsClassifier()
KNN_classifier.fit(X_train, Y_train)
```

```
KNN_scores = cross_val_score(KNN_classifier, X_train, Y_train,
                                scoring="accuracy", cv=10)
```

```
print(KNN_scores.mean())
```
